```python
# lifespans 

# this has to get rid of negative numbers, 
# in future need to make ROIs ffrom top to bottom

lifespans = []
lifespans_ms = []

def get_life(id):
    y0 = df2[id][0][3]
    ymaxind = points_list [id]
    ymax = df2[id][ymaxind][3]
    life_pix = ymax - y0
    life_s = life_pix * strobe_ms
    if life_s > 0:
        lifespans_ms.append(life_s)
    
    else:
        lifespans_ms.append (life_s*-1)
        

        
 
    
for id in trackids:
    get_life(id)
    
for i in lifespans_ms:
        lifespans.append(i/1000)
    
#print (lifespans)
```

```python
# get life of each segment of track


uni_life_ms = []
uni_life_s = []
uni_mid_s = []
uni_mid_ms = []
uni_mid_pix = []

def get_each_life(id):
    ymaxind = points_list [id]+1

    points = list (range(ymaxind))
    ylist = []
    
    for i in points:
        ylist.append (df2[id][i][3])
        
        
    
    uni_life_pix = np.diff(ylist)
    ylen = len(ylist)
    
    y_ids = list(range(1,ylen))
    
    #print (y_ids)
    
    for i in y_ids:
        y0 = ylist[i-1]
        y1 = ylist[i]
        
        ydiff_pix = y1-y0
        ydiff_pix_2 = ydiff_pix/2
        
        mid = y1-ydiff_pix_2
        
        uni_mid_ms.append (mid*int(strobe_ms))
        
        
        #print (mid)
        
    

    
    
    for i in uni_life_pix:
        if i > 0:
            uni_life_ms.append(i*int(strobe_ms))
        else:
            uni_life_ms.append (i*int(strobe_ms)*-1)
            

            
            

    
    #print (ydiff)
        
uni_life_s = []



for id in trackids:
    get_each_life(id)
    
for i in uni_life_ms:
    uni_life_s.append(i/1000)   
    
for i in uni_mid_ms:
    uni_mid_s.append(i/1000)
    

    

#print (uni_mid_s)
```

```python
# hoop length - biggest v smallest x value

hoop_lengths = []

hoop_length_degs = []

def hoop_length (id):
    xmaxind = points_list [id]+1
    points = list (range(xmaxind))
    xlist = []
    points_1 = xmaxind-1
    segs = list (range(1,points_1))
    
    ring_diameter = each_track_diameter[id] 
    
    for i in points:
        xlist.append (df2[id][i][2])
        
        xmin = min(xlist)
        xmax = max(xlist)
        
        hoop_len = xmax-xmin
        
        hoop_len_um = hoop_len*0.065
        
    hoop_lengths.append(hoop_len_um)
    
    
    # hoop length to degrees here 
    
    nm = hoop_len_um *1000
    
    d_nm = ring_diameter
    
    circumference_nm = d_nm * math.pi
    
    nm_per_degree = 360 / circumference_nm 
    
    hoop_len_deg = nm * nm_per_degree
    
    hoop_length_degs.append(hoop_len_deg)
    
    
    
    
    
    
        

    
    
for id in trackids:
    hoop_length(id)
    
#print (hoop_length_degs)
```
